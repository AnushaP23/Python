```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri.head()
```

```python
fmri.shape
```

```python
#Plot a lineplot between timepoint and signal
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri)
plt.show()
```

```python
#event catergory in fmri has 2 categories if we want to plot both that one graph
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event')
plt.show()
```

```python
#Adding style
```

```python
sns.lineplot(x='timepoint',y='signal',hue='event',data=fmri,style='event')
plt.show()
```

```python
#Adding markers
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='event',markers=True)
plt.show()
```

```python
#Seaborn histogram/distplot
```

```python
from matplotlib import pyplot as plt
```

```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
diamonds = pd.read_csv('diamonds.csv')
```

```python
diamonds.head()
```

```python
#Variation of price column
```

```python
sns.distplot(diamonds['price'])
plt.show()
```

```python
#Displot is a combination of histogram and frequency curve
```

```python
#If you want to just display the frequencu curve
```

```python
sns.distplot(diamonds['price'],hist=False)
plt.show()
```

```python
sns.distplot(diamonds['price'],color='purple')
plt.show()
```

```python
#sns.distplot(diamonds['price'],color='purple',kde=False,bins=10)
plt.show()
```

```python
#Vertical plot
```

```python
sns.distplot(diamonds['price'],color='orange',vertical=True)
plt.show()
```

```python
#Seaborn jointplot
```

```python
#Jointplot is a combination of scatterplot and histogram
```

```python
iris.head()
```

```python
sns.jointplot(x='sepal.length',y='petal.length',data=iris)
plt.show()
```

```python
#Up and down is sepal.length and left and right is for petal.length
```

```python
#Changing color of jointplot
```

```python
sns.jointplot(x='sepal.length',y='petal.length',data=iris,color='brown')
plt.show()
```

```python
#kind parameter: is used to add regression line to the graph
```

```python
sns.jointplot(x='sepal.length',y='petal.length',data=iris,color='olive',kind='reg')
plt.show()
```

```python
churn = pd.read_csv('churn.csv')
```

```python
churn.head()
```

```python
sns.boxplot(x='Churn',y='tenure',data=churn)
plt.show()
```

```python
sns.boxplot(x='InternetService',y='MonthlyCharges',data=churn)
plt.show()
```

```python
sns.boxplot(x='InternetService',y='tenure',data=churn,palette='Set1')
plt.show()
```

```python
sns.boxplot(x='InternetService',y='tenure',data=churn,color='blue')
plt.show()
```

```python
sns.boxplot(x='Contract',y='tenure',data=churn,palette='Set1',linewidth=4)
plt.show()
```

```python
sns.boxplot(x='Contract',y='tenure',data=churn,order=['One year','Month-to-month','Two year'])
plt.show()
```

```python
sns.boxplot(x='Contract',y='tenure',data=churn,order=['One year','Month-to-month','Two year'],hue='PaymentMethod')
plt.show()
```
