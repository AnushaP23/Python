```python
#Matplotlib = Matplotlib is a python library used for data visualization
```

```python
#You can create bar-plots, scatter-plots, histogram and a lot more with matplotlib
```

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
```

```python
x
```

```python
y = 3 * x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.show()
```

```python
plt.plot(x,y,color='g',linestyle=':',linewidth=5)
plt.title("X vs Y")
plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.show()
```

```python
#Plotting two lines in one graph
```

```python
x
```

```python
y1 = 2 * x
y2 = 3 * x
```

```python
plt.plot(x, y1, color='r',linestyle=':')
plt.plot(x, y2,color='y',linewidth=4)
plt.title('Two Lines In A Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(True)
plt.show()
```

```python
#Adding subplots
```

```python
plt.subplot(2,1,1)
plt.plot(x, y1, color='r', linestyle=':')
plt.title("Plotting 2 subplots in one screen")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(True)
plt.show()

plt.subplot(2,1,2)
plt.plot(x, y2, color='y', linewidth=4)

plt.title("Plotting 2 subplots in one screen")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(True)
plt.show()
```

```python
#Bar plot is used to understand categorical value distribution
```

```python
student = {'Bob':87, 'Julia':77, 'Anne':100, 'Matt':67}
```

```python
names = list(student.keys())
```

```python
marks = list(student.values())
```

```python
names
```

```python
marks
```

```python
plt.bar(names,marks)
plt.title("Distribution of student marks")
plt.xlabel("Names of students")
plt.ylabel("Marks scored by students")
plt.show()
```

```python
#Horizontal barplots
```

```python
#The only difference is we use barh instead of bar to plot the graph
```

```python
plt.barh(names,marks,color='g')
plt.title('Distribution of student marks')
plt.xlabel('Names of students')
plt.ylabel('Marks scored by students')
plt.show()
```

```python
#Scatter Plot
```

```python
import numpy as np
```

```python
x = [4,2,5,7,2,4,6,2,7,9]
y = [9,3,5,1,8,5,4,3,7,1]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
#To make it more aesthetic
```

```python
plt.scatter(x,y, marker='*',color='r',s=100)
plt.show()
```

```python
#Adding two scatter plots in one plot
```

```python
x = [4,2,5,7,2,4,6,2,7,9]
y = [9,3,5,1,8,5,4,3,7,1]
z = [7,3,4,5,6,1,0,7,3,1]
```

```python
plt.scatter(x,y,color='g',s=200)
plt.scatter(x,z,color='r',marker='*',s=100)
plt.show()
```

```python
#Adding subplots
```

```python
plt.subplot(1,2,1)
plt.scatter(x,y,color='g',marker='*',s=200)
plt.title("1 st plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.subplot(1,2,2)
plt.scatter(x,z,color='r',s=100)
plt.title('2nd plot')
plt.xlabel("X-Axis")
plt.ylabel('Y-Axis')
plt.show()
```

```python
#Histogram
```

```python
#Bar plot is used for categorical data
```

```python
#Histogram is used for numerical data
```

```python
data = [2,2,2,5,5,6,1,7,8,3,6,6,9,2,4,5,6]
```

```python
plt.hist(data,color='r')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title("Histogram")
plt.show()
```

```python
plt.hist(data,color='g',bins=4)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title("Histogram")
plt.show()
```

```python
#Creating histogram on a dataset
```

```python
import pandas as pd
iris = pd.read_csv('iris.csv')
```

```python
iris.head()
```

```python
plt.hist(iris['petal.length'],color='g',bins=30)
plt.title("Histogram on iris dataset(petal.length)")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
#Box-plots
```

```python
#It gives the 5 number summary(min,25 percentile,median,75 percentile,max)
```

```python
one = [1,2,3,4,5,6,7,8,9]
two = [1,2,3,4,5,4,3,2,1]
three = [6,7,8,9,7,6,5,7,8]
```

```python
data = list([one,two,three])
```

```python
data
```

```python
plt.boxplot(data)
plt.show() #if we don't use it, it will plot along with meta-data
```

```python
#Violin plot
```

```python
#Difference between box-plot ans violin plot is the appearance of the plots and rest all is same
#(box-plot looks like box and violin plot looks like violin)
```

```python
plt.violinplot(data,showmedians=True)
plt.show()
```

```python
#Pie-Chart
```

```python
#In pie-chart we can see the distribution value,different percentage of values
```

```python
fruits = ['Apple','Mango','Guava','Orange','Pineapple']
```

```python
quantity = [10,5,15,20,8]
```

```python
plt.pie(quantity, labels=fruits) #First provide the numerical value and then provide the labels
plt.show()
```

```python
#If you want to show percentages and aswell as change the colors
```

```python
plt.pie(quantity, labels='fruit',autopct='%0.1f%%',colors=['yellow','blue','green','red','orange'])
plt.show()
```

```python
#Doughnut-chart
```

```python
#outter graph
plt.pie(quantity,labels=fruits,radius=2)
#inner graph
plt.pie([100],radius=1,colors=['w'])
plt.show()
```
